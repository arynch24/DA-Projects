```python
column_dollar=['price','extra_people','cleaning_fee','security_deposit']
```

```python
# Remove dollar sign
for i in column_dollar:
    df[i] = df[i].str.replace("$", '');  
    
```

```python
dataType=['price','extra_people','cleaning_fee','security_deposit']
```

```python
df[column_dollar].head()
```

```python
for column in dataType:
    # Remove commas and other non-numeric characters (like '$')
    df[column] = df[column].replace({'\$': '', ',': ''}, regex=True)
    
    # Convert to float64
    df[column] = df[column].astype('float64')
```

```python
df.head()
```

```python
num=[]
cat=[]
```

```python
for i in df:
    if(df[i].dtype =='int64' or df[i].dtype=='float64'):
        num.append(i)
    else:
        cat.append(i)
```

```python
len(num)
```

```python
len(cat)
```

```python
df[num].head()
```

```python
df[cat].head()
```

```python
##null value treatment
#for categorical value

for i in cat:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
#check
df[cat].isnull().sum()
```

```python
#treating null values of num
for i in num:
    df[i]=df[i].fillna(df[i].median())
```
